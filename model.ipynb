```python
from sklearn.preprocessing import LabelEncoder
```

```python
Fat_Content_label_encoder_1 = LabelEncoder()
df['Fat_content_encoder'] = Fat_Content_label_encoder_1.fit_transform(df['Item_Fat_Content'])
```

```python
Item_Type_label_encoder_2 = LabelEncoder()
df['Item_type_encoder'] = Item_Type_label_encoder_2.fit_transform(df['Item_Type'])
```

```python
Identifier_label_encoder_3 = LabelEncoder()
df['Outlet_Identifier_enocder'] = Identifier_label_encoder_3.fit_transform(df['Outlet_Identifier'])
```

```python
Establishment_Year_encoder_4 = LabelEncoder()
df['Establishment_Year_encoder'] = Establishment_Year_encoder_4.fit_transform(df['Outlet_Establishment_Year'])
```

```python
Size_encoder_5 = LabelEncoder()
df['Size_encoder'] = Size_encoder_5.fit_transform(df['Outlet_Size'])
```

```python
df['Size_encoder']
```

```python
Location_Type_encoder_6 = LabelEncoder()
df['Location_type_encoder']  = Location_Type_encoder_6.fit_transform(df['Outlet_Location_Type'])
```

```python
Outlet_Type_encoder_7 = LabelEncoder()
df['Outlet_type'] = Outlet_Type_encoder_7.fit_transform(df['Outlet_Type'])
```

```python
x = df[['Fat_content_encoder','Item_type_encoder','Outlet_Identifier_enocder','Establishment_Year_encoder','Size_encoder','Location_type_encoder','Outlet_type']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(x_train, y_train)
```
